## Tight binding model

```python
def Ham0(N):
  H=np.zeros([N,N])
  for i in range(N):
    H[i,(i+1)%N]=-1
    H[(i+1)%N,i]=-1
  return H
plt.matshow(Ham0(11))
print(np.transpose(Ham0(11))==Ham0(11))
```

```python
def FSpace0(N):
  matr=np.zeros([N,N],dtype = 'complex_')
  for i in range(N):
    for j in range(N):
      k=MLTK0((j,'a'),N,1)
      r=LTR0((i,'a'))
      matr[i,j]=1/np.sqrt(N)*np.exp(1j*k*r)
  return matr
FMat = FSpace0(5)
# print(FMat @ FMat.conjugate().transpose())
# print(np.FMat)
```

```python
def LTI2(label, N1, N2):
  if label[1]=='a':
    return label[0][0]*N2*2+label[0][1]*2
  else:
    return label[0][0]*N2*2+label[0][1]*2+1
labels = Labels2(3,4)
```

```python
import cmath
def BF(x,y,N1,N2,m, H):
  locs = [(x,y),((x+1)%N1,y),((x+1)%N1,(y-1)%N2),(x,(y-1)%N2)]
  vs=[]
  for loc in locs:
    i = loc[0]*loc[1]
    Hblock = np.array([[H[2*i][2*i],H[2*i][2*i+1]],[H[2*i+1][2*i],H[2*i+1][2*i+1]]])
    w,v1 = np.linalg.eigh(Hblock)
    vs.append(np.array(v1[:,0]))
  v = (vs[0].conj().T @ vs[1])*(vs[1].conj().T @ vs[2])*(vs[2].conj().T @ vs[3])*(vs[3].conj().T @ vs[0])
  return cmath.phase(v)
```
